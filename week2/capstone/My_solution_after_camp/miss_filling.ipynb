```python
import pandas as pd
from helper import analyze_data, knn_imputation, visual_miss
```

### broadband

```python
broadband = pd.read_csv("../Data/broadband.csv")
filled_broadband = broadband.copy()
```

```python
broadband.head()
```

```python
broadband.info()
```

```python
visual_miss(broadband)
```

```python
cat_cols, num_cols, cat_but_car = analyze_data(broadband)
```

```python
broadband[cat_cols].isna().sum()
```

```python
filled_broadband["auto_payment"] = knn_imputation(filled_broadband, categorical_columns=["auto_payment"])["auto_payment"]
```

```python
broadband[num_cols].isna().sum()
```

```python
mis_num_cols = ["tenure", "data_usage", "monthly_charge"]
filled_broadband[mis_num_cols] = knn_imputation(filled_broadband, numerical_columns=mis_num_cols)[mis_num_cols]
```

```python
filled_broadband.isna().sum()
```

```python
filled_broadband.to_csv("../Data/filled_broadband.csv", index=False)
```
