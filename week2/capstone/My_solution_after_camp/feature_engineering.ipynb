#### 1. Kategori Dönüşümleri

```python
broadband['age_group'] = pd.cut(broadband['age'], 
                               bins=[0, 25, 35, 45, 60, 100], 
                               labels=['18-25', '26-35', '36-45', '46-60', '60+'])
```

```python
broadband['customer_segment'] = pd.cut(broadband['tenure'], 
                                      bins=[0, 12, 24, 36, 120, 240, np.inf], 
                                      labels=['New', '1-2 Year', '2-3 Year', '3-10 Year', '10-20 Year', '20+ Year'])
```

```python
broadband["data_usage_group"] = pd.cut(broadband["data_usage"],
                                      bins=[0, 10, 20, 50, 100, np.inf],
                                      labels=['0-10', '10-20', '20-50', '50-100', '100+'])
```

```python
broadband["monthly_charge_group"] = pd.cut(broadband["monthly_charge"],
                                          bins=[0, 200, 250, 300, 400, np.inf],
                                          labels=['0-200', '200-250', '250-300', '300-400', '400+'])
```

```python
broadband["satisfaction_score_group"] = pd.cut(broadband["satisfaction_score"],
                                              bins=[0, 2, 4, 6, 8, 10],
                                              labels=['0-2', '2-4', '4-6', '6-8', '8-10'])
```

#### 2. Oranlar

```python
broadband['data_per_dollar'] = broadband['data_usage'] / broadband['monthly_charge']
```

```python
broadband['support_interaction_ratio'] = broadband['customer_support_calls'] / broadband['tenure']
```

#### 3. Etkileşim Terimleri

##### Memnuniyet skoru ve çağrı sayısı arasındaki etkileşim

```python
broadband['call_satisfaction_interaction'] = broadband['customer_support_calls'] * broadband['satisfaction_score']
```

##### Vadesi geçmiş ödemeler ve otomatik ödeme arasındaki etkileşim

```python
broadband['payment_issues'] = broadband['overdue_payments'] * (1 - broadband['auto_payment'])
```

#### 4. Zaman Tabanlı Özellikler

##### Aylık maliyet / kullanım süresi - dakika başına maliyet

```python
broadband['monthly_charge_per_tenure'] = broadband['monthly_charge'] / (broadband['tenure'] + 1)
```

##### Ay cinsinden yaş

```python
broadband["monthly_age"] = broadband["age"] * 12
```

##### Müşteri olduğu yaş

```python
broadband["come_age"] = broadband["tenure"] - broadband["monthly_age"]
```

#### 5. Agregasyonlar ve İstatistikler

##### Veri kullanım yoğunluğu (ortalama kullanıma göre normalize edilmiş)

```python
broadband['data_usage_intensity'] = broadband['data_usage'] / broadband['data_usage'].mean()
```

##### Aykırı değer tespiti

```python
broadband['is_heavy_user'] = (broadband['data_usage'] > broadband['data_usage'].quantile(0.9)).astype(int)
broadband['is_expensive_plan'] = (broadband['monthly_charge'] > broadband['monthly_charge'].quantile(0.9)).astype(int)
```

#### 6. Müşteri Memnuniyeti İndeksi

##### Memnuniyet, ters çağrı sayısı ve ters vadesi geçmiş ödemelerden oluşan bileşik skor

```python
max_calls = broadband['customer_support_calls'].max()
max_overdue = broadband['overdue_payments'].max()
```

##### Normalize ve birleştir (her biri 0-1 aralığında)

```python
broadband['customer_health_index'] = (
    (broadband['satisfaction_score'] / 5) * 0.5 +
    (1 - broadband['customer_support_calls'] / (max_calls + 1)) * 0.3 +
    (1 - broadband['overdue_payments'] / (max_overdue + 1)) * 0.2
)
```

#### 7. Risk Skoru

```python
broadband['churn_risk_score'] = (
    (broadband['overdue_payments'] * 0.3) +
    (broadband['customer_support_calls'] * 0.3) +
    ((6 - broadband['satisfaction_score']) * 0.4)
) / 10
```

#### 8. Davranışsal Özellikler

##### Ödeme davranışı (otomatik ödeme ve vadesi geçmiş ödemeler birleşimi)

```python
broadband['payment_behavior'] = np.where(
    (broadband['auto_payment'] == 1) & (broadband['overdue_payments'] == 0), 'Çok İyi',
    np.where((broadband['auto_payment'] == 1) & (broadband['overdue_payments'] > 0), 'İyi',
            np.where((broadband['auto_payment'] == 0) & (broadband['overdue_payments'] <= 2), 'Orta',
                    'Kötü'))
)
```

##### Müşteri destek bağımlılığı

```python
broadband['support_dependency'] = broadband['customer_support_calls'] / (broadband['tenure'] + 1)
```

##### 9. Logaritmik ve Karekök Dönüşümleri

##### Çarpık değişkenler için logaritmik dönüşüm

```python
broadband['log_data_usage'] = np.log1p(broadband['data_usage'])
broadband['log_tenure'] = np.log1p(broadband['tenure'])
```

#### 10. Uygulamalar

```python
app_cols = ['İzleGo', 'RitimGo', 'CüzdanX', 'HızlıPazar', 'Konuşalım']
broadband['total_apps_used'] = broadband[app_cols].sum(axis=1)
```
